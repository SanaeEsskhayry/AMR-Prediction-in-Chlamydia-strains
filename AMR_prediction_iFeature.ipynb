Concatenate AMR features

```python
import pandas as pd
# List of file paths to the TSV files
file_paths = ["AMR_Features/AAC.tsv", "AMR_Features/CKSAAP.tsv", "AMR_Features/CTDC.tsv", "AMR_Features/CTDD.tsv", "AMR_Features/CTDT.tsv", "AMR_Features/CTriad.tsv","AMR_Features/DDE.tsv","AMR_Features/DPC.tsv","AMR_Features/GAAC.tsv","AMR_Features/KSCTriad.tsv","AMR_Features/GDPC.tsv","AMR_Features/CKSAAGP.tsv","AMR_Features/GTPC.tsv"]

# List to hold individual DataFrames
dataframes = []

# Loop through each file path and read the TSV file into a DataFrame
for file_path in file_paths:
    df = pd.read_csv(file_path, sep="\t")
    print(df.shape)
    df = df.drop(df.columns[0], axis=1)
    dataframes.append(df)

AMR_Features_data = pd.concat(dataframes, axis=1)
```

Concatenate Sensitive features

```python
import pandas as pd
# List of file paths to the TSV files
file_paths = ["Sensitive_features/Sen_AAC.tsv", "Sensitive_features/Sen_CKSAAP.tsv", "Sensitive_features/Sen_CTDC.tsv", "Sensitive_features/Sen_CTDD.tsv", "Sensitive_features/Sen_CTDT.tsv", "Sensitive_features/Sen_CTriad.tsv","Sensitive_features/Sen_DDE.tsv","Sensitive_features/Sen_DPC.tsv","Sensitive_features/Sen_GAAC.tsv","Sensitive_features/Sen_KSCTriad.tsv","Sensitive_features/Sen_GDPC.tsv","Sensitive_features/Sen_CKSAAGP.tsv","Sensitive_features/Sen_GTPC.tsv"]

# List to hold individual DataFrames
dataframes = []

# Loop through each file path and read the TSV file into a DataFrame
for file_path in file_paths:
    df = pd.read_csv(file_path, sep="\t")
    print(df.shape)
    df = df.drop(df.columns[0], axis=1)
    dataframes.append(df)

# Concatenate the DataFrames along rows (vertically)
Sen_Features_data = pd.concat(dataframes, axis=1)
```

Calculate feature for both positive and negative classes + combines the two classes + merge with class labels

```python
# Create class labels
Sen_class = pd.Series(['Sensitive' for i in range(len(Sen_Features_data))])
AMR_class = pd.Series(['Resistant' for i in range(len(AMR_Features_data))])
# Combine AMR Data and Sensitive Data
AMR_Sen_class = pd.concat([Sen_class, AMR_class], axis=0)
AMR_Sen_class.name = 'class'
AMR_Sen_feature = pd.concat([Sen_Features_data, AMR_Features_data], axis=0)
# Combine feature and class
df = pd.concat([AMR_Sen_feature, AMR_Sen_class], axis=1)
```
