Secondary Structure Elements Content (SSEC)

```python
from Bio import SeqIO

def read_fasta_sequences(file_path):
    sequences = []
    with open(file_path, "r") as fasta_file:
        for record in SeqIO.parse(fasta_file, "txt"):
            sequence_name = record.id
            sequence_data = str(record.seq)
            sequences.append((sequence_name, sequence_data))
    return sequences

# Replace 'your_input_file.fasta' with the actual path to your FASTA file
sequences = read_fasta_sequences('AMR_sequences_cdhit.txt')
print(sequences)

```
